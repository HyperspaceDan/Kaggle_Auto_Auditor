# Strings


One place where the Python language really shines is in the manipulation of strings.
This section will cover some of Python's built-in string methods and formatting operations.

Such string manipulation patterns come up often in the context of data science work.

## String syntax

You've already seen plenty of strings in examples during the previous lessons, but just to recap, strings in Python can be defined using either single or double quotations. They are functionally equivalent. While both are valid, it's a good practice (e.g., following [PEP 8](https://www.python.org/dev/peps/pep-0008/#string-quotes)) to pick a consistent style for your projects, often using single quotes for simple strings and double quotes when the string itself contains single quotes.


```python
x = 'Pluto is a planet'
y = "Pluto is a planet"
x == y
```

Double quotes are convenient if your string contains a single quote character (e.g. representing an apostrophe).

Similarly, it's easy to create a string that contains double-quotes if you wrap it in single quotes:


```python
print("Pluto's a planet!")
print('My dog is named "Pluto"')
```

If we try to put a single quote character inside a single-quoted string *without escaping it*, Python gets confused and will raise a **`SyntaxError`** because it interprets the inner quote as the end of the string. For example, the next cell will cause an error:


```python
# This line will cause a SyntaxError because the single quote in "Pluto's"
# prematurely terminates the string literal.
# Uncommenting the line below will raise a SyntaxError:
# 'Pluto's a planet!'

# Correct ways to handle this are shown in the previous cell (using double quotes)
# or in the next cell (by escaping the single quote).
```

We can fix this by "escaping" the single quote with a backslash. 


```python
'Pluto\'s a planet!'
```

The table below summarizes some important uses of the backslash character.

| What you type... | What you get | example               | `print(example)`             |
|--------------|----------------|--------------------------------------------------------|
| `\'`         | `'`            | `'What\'s up?'`         | `What's up?`                 |  
| `\"`         | `"`            | `"That's \"cool\""`     | `That's "cool"`              |  
| `\\`         | `\`            |  `"Look, a mountain: /\\"` |  `Look, a mountain: /\`  |
| `\n`        |   <br/>      |   `"1\n2 3"`                       |   `1`<br/>`2 3`              |


The last sequence, `\n`, represents the *newline character*. It causes Python to start a new line.


```python
hello = "hello\nworld"
print(hello)
```

In addition, Python's triple quote syntax for strings lets us include newlines literally (i.e. by just hitting 'Enter' on our keyboard, rather than using the special '\n' sequence). We've already seen this in the docstrings we use to document our functions, but we can use them anywhere we want to define a string.


```python
triplequoted_hello = """hello
world"""
print(triplequoted_hello)
triplequoted_hello == hello
```

The `print()` function automatically adds a newline character unless we specify a value for the keyword argument `end` other than the default value of `'\n'`:


```python
print("hello")
print("world")
print("hello", end='')
print("pluto", end='')
```

## Strings are sequences

Strings can be thought of as sequences of characters, much like lists. This means they support operations like **indexing** and **slicing** to access parts of the string, and you can even **loop** over them.


```python
# Indexing
planet = 'Pluto'
planet[0]
```

```python
# Slicing
planet[-3:]
```

```python
# How long is this string?
len(planet)
```

```python
# Yes, we can even loop over them
[char+'! ' for char in planet]
```

But a major way in which they differ from lists is that they are **immutable**. This means that once a string is created, its individual characters cannot be changed. Attempting to modify a string by index, like `planet[0] = 'B'`, will result in a **`TypeError`** because string objects do not support item assignment. To "change" a string, you must create a *new* string based on the old one.


```python
# This line will cause a TypeError because strings are immutable.
# You cannot change individual characters in a string after it's created.
# Uncommenting the line below will raise a TypeError:
# planet[0] = 'B'

# To "change" a string, you must create a new one based on the original.
# For example:
new_planet = 'B' + planet[1:]
print(f"Original planet: {planet}")
print(f"New planet (first char changed): {new_planet}")

# As the comment below correctly states, string objects also don't have an 'append' method.
# planet.append doesn't work either
```

## String methods

Like `list`, the type `str` has many useful methods for common operations. Here are some examples, including those for case conversion, searching, and checking prefixes/suffixes.


```python
# ALL CAPS
claim = "Pluto is a planet!"
claim.upper()
```

```python
# all lowercase
claim.lower()
```

```python
# Searching for the first index of a substring
# If the substring is not found, .index() will raise a ValueError.
print(f"Index of 'plan': {claim.index('plan')}")

# An alternative, .find(), returns -1 if the substring is not found,
# which can be useful if you just want to check for existence without
# handling an exception.
print(f"Index of 'planet': {claim.find('planet')}")
print(f"Index of 'xyz' (not found): {claim.find('xyz')}")

# Uncommenting the line below would raise a ValueError:
# claim.index('xyz')
```

```python
claim.startswith(planet)
```

```python
# false because of missing exclamation mark
claim.endswith('planet')
```

### Going between strings and lists: `.split()` and `.join()`

`str.split()` turns a string into a list of smaller strings, breaking on whitespace by default. This is super useful for taking you from one big string to a list of words.


```python
words = claim.split()
words
```

Occasionally you'll want to split on something other than whitespace:


```python
datestr = '1956-01-31'
year, month, day = datestr.split('-')
```

`str.join()` takes us in the other direction, sewing a list of strings up into one long string, using the string it was called on as a separator.


```python
'/'.join([month, day, year])
```

```python
# Yes, we can put unicode characters right in our string literals :)
' 👏 '.join([word.upper() for word in words])
```

### Building strings with `.format()` and f-strings

Python offers several powerful ways to build strings, especially when you need to embed variables or format values. Let's start with simple concatenation using the `+` operator.


```python
planet + ', we miss you.'
```

When concatenating strings using the `+` operator, all operands must be strings. If you try to concatenate a string with a non-string object (like an integer), Python will raise a **`TypeError`**. You must explicitly convert non-string objects to strings using `str()` first.


```python
position = 9
# This line will cause a TypeError because 'position' is an integer and cannot be
# directly concatenated with strings using the '+' operator.
# Uncommenting the line below will raise a TypeError:
# planet + ", you'll always be the " + position + "th planet to me."

# The next cell demonstrates the correct way to handle this.
```

```python
planet + ", you'll always be the " + str(position) + "th planet to me."
```

This is getting hard to read and annoying to type. Python offers much cleaner ways to build strings, primarily using `str.format()` and, more recently, f-strings (formatted string literals).


```python
"{}, you'll always be the {}th planet to me.".format(planet, position)
```

So much cleaner! We call `.format()` on a "format string", where the Python values we want to insert are represented with `{}` placeholders.

Notice how we didn't even have to call `str()` to convert `position` from an int. `format()` takes care of that for us.

If that was all that `format()` did, it would still be incredibly useful. But as it turns out, it can do a *lot* more. Here's just a taste of its advanced capabilities:


```python
pluto_mass = 1.303 * 10**22
earth_mass = 5.9722 * 10**24
population = 52910390
#         2 decimal points   3 decimal points, format as percent     separate with commas
"{} weighs about {:.2} kilograms ({:.3%} of Earth's mass). It is home to {:,} Plutonians.".format(
    planet, pluto_mass, pluto_mass / earth_mass, population,
)
```

```python
# Referring to format() arguments by index, starting from 0
s = """Pluto's a {0}.
No, it's a {1}.
{0}!
{1}!""".format('planet', 'dwarf planet')
print(s)
```

### F-strings (Formatted String Literals)

Introduced in Python 3.6, f-strings provide a more concise and readable way to embed expressions inside string literals. They are prefixed with `f` or `F` and use curly braces `{}` to contain expressions that will be evaluated at runtime and then formatted.

F-strings offer the same powerful formatting capabilities as `str.format()`, but often with less boilerplate.


```python
# Simple f-string example
f"{planet}, you'll always be the {position}th planet to me."
```

```python
# F-string with advanced formatting, mirroring the .format() example
pluto_mass = 1.303 * 10**22
earth_mass = 5.9722 * 10**24
population = 52910390
f"{planet} weighs about {pluto_mass:.2} kilograms ({pluto_mass / earth_mass:.3%} of Earth's mass). It is home to {population:,} Plutonians."
```

```python
# Multi-line f-string with expressions
planet_type_1 = 'planet'
planet_type_2 = 'dwarf planet'
s_fstring = f"""Pluto's a {planet_type_1}.
No, it's a {planet_type_2}.
{planet_type_1}!
{planet_type_2}!"""
print(s_fstring)
```

You could probably write a short book just on string formatting! For `str.format()`, I'll stop here and point you to [pyformat.info](https://pyformat.info/) and [the official docs](https://docs.python.org/3/library/string.html#formatstrings) for further reading. For f-strings, the [official Python documentation](https://docs.python.org/3/reference/lexical_analysis.html#f-strings) is an excellent resource.


# Dictionaries

Dictionaries are a built-in Python data structure for mapping keys to values.


```python
numbers = {'one':1, 'two':2, 'three':3}
```

In this case `'one'`, `'two'`, and `'three'` are the **keys**, and 1, 2 and 3 are their corresponding values.

Values are accessed via square bracket syntax similar to indexing into lists and strings.


```python
numbers['one']
```

We can use the same syntax to add another key, value pair


```python
numbers['eleven'] = 11
numbers
```

Or to change the value associated with an existing key


```python
numbers['one'] = 'Pluto'
numbers
```

Python has *dictionary comprehensions* with a syntax similar to the list comprehensions we saw in the previous tutorial.


```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planet_to_initial = {planet: planet[0] for planet in planets}
planet_to_initial
```

The `in` operator tells us whether something is a key in the dictionary


```python
'Saturn' in planet_to_initial
```

```python
'Betelgeuse' in planet_to_initial
```

A for loop over a dictionary will loop over its keys


```python
for k in numbers:
    print("{} = {}".format(k, numbers[k]))
```

We can access a collection of all the keys or all the values with `dict.keys()` and `dict.values()`, respectively.


```python
# Get all the initials, sort them alphabetically, and put them in a space-separated string.
' '.join(sorted(planet_to_initial.values()))
```

The very useful `dict.items()` method lets us iterate over the keys and values of a dictionary simultaneously. (In Python jargon, an **item** refers to a key, value pair)


```python
for planet, initial in planet_to_initial.items():
    print("{} begins with \"{}\"".format(planet.rjust(10), initial))
```
